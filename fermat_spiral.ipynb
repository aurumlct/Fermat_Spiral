# Fermat spiral

## I. Floret positions

1. Importing modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

2. Define function to convert polar to Cartesian coordinates

```python
def polar2cart(r, theta):
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    return x, y
```

3. Specify number of florets to plot

```python
nmax = 20
```

4. Parameter to alter positions of florets (=0 for positions that line on a Fermat spiral)

```python
ndiff = 0
```

5. Convert to Cartesian

```python
def floret_carts(nmax, ndiff):
    n = np.arange(1,nmax+1)
    rf = np.sqrt(n)
    phi = (1 + np.sqrt(5))/2
    thetaf = (2*np.pi/phi**2+ndiff)*n
    xf, yf = polar2cart(rf, thetaf)
    return xf, yf

# Function call
xf, yf = floret_carts(nmax, ndiff)
```

6. Plot floret positions

```python
plt.plot(xf,yf,'ro')
plt.title('Fermat spiral', color='blue', fontname='Comic Sans MS')
```

## II. Fermat spiral

1. Define the Golden Mean

```python
phi = (1 + np.sqrt(5))/2
```

2. Define theta and r

```python
theta = np.linspace(0,50,5000)
r = np.sqrt(phi**2*theta/(2*np.pi))
```

3. Calling polar2cart function to convert theta and r

```python
x, y = polar2cart(r, theta)
```

4. Ploting Fermat spiral on top of florets

```python
plt.plot(xf,yf,'ro')
plt.plot(x,y)
plt.title('Fermat spiral', color='blue', fontname='Comic Sans MS')
```

## Extra: Making subplots of florets with different nDiff values

```python
nDiff = np.arange(-.25, .26, .1)
figure = plt.figure(figsize=(12,7.8), dpi=100)
for (i, nDiff) in enumerate(nDiff, start=1):
    print(i, nDiff)
    figure.add_subplot(2,3,i)
    plt.title("nDiff={:4.2f}".format(nDiff), fontname='Comic Sans MS', color='orange', fontsize=13)
    xF, yF = floret_carts(nmax, nDiff)
    plt.plot(xF, yF, "bo")

plt.show()
```
